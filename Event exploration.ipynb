```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',None)
```

```python
df = pd.read_csv('Resources/eventfeedbackcleanest.csv')
df
```

```python
df=df.dropna(subset='Rebook Rating')
```

```python
df=df.drop(df.columns[[0,1,6,8,11,12,13,14,15,16,17]], axis=1)
```

```python
df=df.dropna()
```

```python
df.reset_index()
```

```python
from pathlib import Path  
filepath = Path('Resources/out.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(filepath) 
```

```python
df
```

```python
X = df.drop(['Rebook Rating'], axis=1)
y = df['Rebook Rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.1, random_state=42)
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
reg = LinearRegression().fit(X_train_scaled, y_train)
reg.score(X_test_scaled, y_test)
```

```python
base_estimator = DecisionTreeClassifier(random_state=42)
```

```python
bagging_model = BaggingClassifier(base_estimator=base_estimator, n_estimators=10, random_state=42)
```

```python
bagging_model.fit(X_train, y_train)
```

```python
score = bagging_model.score(X_test, y_test)
print("Accuracy:", score)
```

```python
decision_tree = DecisionTreeClassifier(random_state=42)
```

```python
decision_tree.fit(X_train, y_train)
```

```python
importances = decision_tree.feature_importances_
print("Feature importances:", importances)
```
